```python
import pandas as pd #Using Pandas for dataframe
```

```python
coin_list=['BTC','USDC','ALGO','AAVE','FIL','ETC','GRT','LTC','BCH','LINK','ETH','UNI','DOT','XLM','DOGE','USDT','ADA','XRP','EOS']
# Coin list obtained.
```

```python
myVars = vars()
#converting string into variable
```

```python
pd.options.display.float_format = '{:.4f}'.format
```

```python
Dataframe_to_be_push=pd.DataFrame()
```

```python
import requests
#Importing request to fetch live data of Crypto currency
```

```python
for i in coin_list:
    print(i)
    x='https://min-api.cryptocompare.com/data/v2/histoday?fsym=Replace&tsym=USD&limit=2000'
    link = x.replace("Replace", i )
    s=requests.get(link)
    data=s.json()
    df = pd.DataFrame(data['Data']['Data'])
    df['time'] = pd.to_datetime(df['time'], unit='s')
    df["Symbol"]=i
    df2=df['Symbol']
    df.drop(['Symbol'],axis=1,inplace=True)
    df.drop(['conversionSymbol'],axis=1,inplace=True)
    df.drop(['conversionType'],axis=1,inplace=True)
    df.rename(columns={'time':'Time','high':'High','low':'Low','open':'Open','volumefrom':'Starting Day Volume','volumeto':'Closing Day Volume','close':'Close'},inplace=True)
    df=pd.concat([df2,df],axis=1)
    myVars[i] = df.copy()
    Dataframe_to_be_push=pd.concat([Dataframe_to_be_push,df],axis=0)
```
